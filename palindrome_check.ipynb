# Project 11: Check if a String is a Palindrome

```python
# This project teaches how to reverse a string and compare it
```

```python
# Step 1: Ask the user for a string
user_text = input("Enter a text: ")
```

```python
# Step 2: Reverse the string
reversed_text = user_text[::-1]
```

```python
# Step 3: Compare the original and reversed string
if user_text == reversed_text:
    print(f"'{user_text}' is a palindrome!")
else:
    print(f"'{user_text}' is NOT a palindrome.")
```
